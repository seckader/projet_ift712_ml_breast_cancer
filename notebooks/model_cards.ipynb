```python
from pathlib import Path
import pandas as pd
from IPython.display import Image, display

REPORTS = Path("../models/reports")

def show_model_card(model_name: str):
    print(f"=== {model_name.upper()} ===")

    metrics = pd.read_csv(REPORTS / f"{model_name}_test_metrics.csv")
    display(metrics)

    cv = pd.read_csv(REPORTS / f"{model_name}_cv_results.csv")
    best_row = cv.loc[cv["rank_test_score"].idxmin()]
    print("Best CV score:", best_row["mean_test_score"])
    print("Best params:", best_row["params"])

show_model_card("knn")
```

```python
display(Image(filename=str(REPORTS / "knn_confusion_matrix.png")))
```

```python
report_txt = (REPORTS / "knn_classification_report.txt").read_text(encoding="utf-8")
print(report_txt)
```
