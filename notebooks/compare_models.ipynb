```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

REPORTS = Path("../models/reports")

summary_path = REPORTS / "summary.csv"
summary = pd.read_csv(summary_path)

summary
```

```python
f1_summary_sorted = summary.sort_values("f1_macro", ascending=True)

plt.figure()
plt.barh(f1_summary_sorted["model"], f1_summary_sorted["f1_macro"])
plt.title("Comparaison des mod√®les (F1-macro)")
plt.xlabel("F1-macro")
plt.tight_layout()
plt.show()
```
