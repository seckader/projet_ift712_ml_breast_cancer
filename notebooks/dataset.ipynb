## Dataset : Breast Cancer Wisconsin (Diagnostic)

- Nombre d'observations : 569
- Nombre de variables : 30 variables numériques
- Cible : `diagnosis`
  - M = Malignant
  - B = Benign


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/breast_cancer.csv")
df.head()
```

```python
df["target"].value_counts(normalize=True)
```

```python
corr = df.drop(columns=["target"]).corr()
plt.figure(figsize=(10,8))
plt.imshow(corr)
plt.title("Matrice de corrélation (features)")
plt.colorbar()
plt.tight_layout()
plt.show()
```

- Les variables sont toutes numériques.
- Certaines variables semblent fortement discriminantes entre classes.
- Aucune valeur manquante n’est observée.
- Ces observations justifient l’utilisation de modèles classiques supervisés.
